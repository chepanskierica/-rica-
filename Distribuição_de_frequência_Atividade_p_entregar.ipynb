##                                            **TEXTE DE ADERÊNCIA VISUAL**














```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import chi2
```

```
QK=np.arange(0,6,1);
FO=np.array([15,22,15,4,3,1]);
FT=np.array([15.55,21.00,14.17,6.38,2.15,0.58]);
plt.plot(QK,FO,'b-o', QK,FT,'r-*')
plt.title("TESTE DE ADERÊNCIA VISUAL");
plt.legend(['FEQUENCIA OBSERVADA', 'FREQUENCIA CALCULADA'],loc=1)
plt.xlabel('Frequencia Calculada')
plt.ylabel('Frequencia Observada')
plt.show()
```

**TEXTE QUI-QUADRADO**

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import chi2
```

```
Qui2=sum((FO-FT)**2/FT); #QUI-QUADRADO 
k=len(FO); #NÚMERO DE CLASSES
p=0.95 #PROBABILIDADE
c=1; #PARÂMETRO
df=k-1-c #GRAUS DE LIBERDADE
QuiCritical=chi2.ppf(p,df)
print(QuiCritical)
print(Qui2)
```

**DECISÃO E CONCLUSÃO**

```
if (Qui2 < QuiCritical):
  print('Como o Qui2 é < QuiCritical não se pode rejeitar o H0')
else:
  print('Como o Qui2 é > QuiCritical rejeita-se o H0 ')
```
