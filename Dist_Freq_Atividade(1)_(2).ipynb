```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import chi2
```

===========================TESTE DE ADERÊNCIA VISUAL=======================

```
QK=np.arange(0, 14, 1);

FO =np.array([3,15,47,76,68,74,46,39,15,9,5,2,0,1]);
FT=np.array([4,18.4,42.8,65.2,74.8,69.2,52.8,34.8,20,12,4.8,2,0.8,0.4]);
plt.plot(QK, FO ,'b-o', QK, FT, 'r-*')
plt.title("TESTE DE ADERÊNCIA VISUAL")
plt.legend(['FREQUÊNCIA OBSERVADA', 'FREQUÊNCIA TEÓRICA'], loc=1)
plt.xlabel('CLASSES')
plt.ylabel('FREQUÊNCIAS')
plt.show();
```

==================================TESTE DO QUI-QUADRADO====================

```
Qui2=sum((FO-FT)**2/FT); #QUI-QUADRADO 
k=len(FO); #NÚMERO DE CLASSES
p=0.95 #PROBABILIDADE
c=1; #PARÂMETRO
df=k-1-c #GRAUS DE LIBERDADE
QuiCritical=chi2.ppf(p,df)
#print(Qui2)
#print(QuiCritical)
```

===========================DECISÃO E CONCLUSÃO===========================

```
if (Qui2 < QuiCritical):
  print('Como o Qui2 é < QuiCritical não se pode rejeitar o H0')
else:
  print('Como o Qui2 é > QuiCritical rejeita-se o H0 ')
```
